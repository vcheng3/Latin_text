```python
lit = {}
for x in output:
    for i in x[1]:
        if d.get(i[0]) == None:
            p = i[0]
            if d.get(i[1])== None:
                l = i[1]
                j = (p,l)
                if lit.get(str(j)) == None:
                    lit[str(j)]=[x[0]]
                else:
                    lit.get(str(j)).append(x[0])
            else:
                l = d.get(i[1])
                for m in range(len(l)):
                    j = (p,l[m])
                    if lit.get(str(j)) == None:
                        lit[str(j)]=[x[0]]
                    else:
                        lit.get(str(j)).append(x[0])
        else:
            p = d.get(i[0])
            for m in range(len(p)):
                if d.get(i[1])== None:
                    j = (p[m],i[1])
                    if lit.get(str(j)) == None:
                        lit[str(j)]=[x[0]]
                    else:
                        lit.get(str(j)).append(x[0])
                else:
                    l = d.get(i[1])
                    for k in range(len(l)):
                        j = (p[m],l[k])
                        if lit.get(str(j)) == None:
                            lit[str(j)]=[x[0]]
                        else:
                            lit.get(str(j)).append(x[0])
```
